**4a.** (5 pts) Define a function that returns an exponential decay plus a constant $A e^{-t/\tau} + C$. Input parameters to the funciton should consist of a time array followed by any other needed parameters (e.g. time constant, etc.).

```python
def expDecay(time, amp, tau, constant):
    return amp * np.exp(-time / tau) + constant
```

**4b.** (10 pts) Use `scipy.optimize.curve_fit` to optimize the parameters to your function above by minimizing the sum of squared errors between the data given above and the fit returned by your function. Report the optimized parameters along with their units.

```python
import scipy.optimize as opt
result = opt.curve_fit(expDecay, time_sec, voltage_mV)
amp, tau, constant = result[0]
print(f"amp = {amp:.0f} mV, tau = {tau:.1f} sec, constant = {constant:0.1f} mV")
```

**4c.** (5 pts) Plot the data overlaid with the fit. Label all axes and units and add a legend for the data and the fit.

```python
plt.plot(time_sec, voltage_mV, label='data')
plt.plot(time_sec, expDecay(time_sec, amp, tau, constant), lw=3, label='fit')
plt.xlabel('Time (sec)')
plt.ylabel('Voltage (mV)')
plt.legend();
```

**6e.** (10 pts) Test your null hypothesis using a permutation test with 10,000 permutations. Plot a histogram of the permuted differences in standard deviation with 100 bins. Plot dashed vertical lines for the 95% confidence interval of your permuted distribution. Also plot a vertical line for the measured difference in standard deviation (from the data given above).

```python
na = len(control_pA)
nb = len(toxin_pA)
allcurrents = np.zeros((na+nb,))
allcurrents[:na] = control_pA
allcurrents[-nb:] = toxin_pA
permstats = np.zeros((10000))
for i in range(10000):
    np.random.shuffle(allcurrents)
    perma = allcurrents[:na]
    permb = allcurrents[-nb:]
    permstats[i] = permb.std() - perma.std()

plt.hist(permstats, bins=100, alpha=0.5)
plt.xlabel('Current Stdev (Toxin - Control)')
plt.ylabel('# Permutations');

lb, ub = np.quantile(permstats, [0.025, 0.975])
plt.axvline(lb, linestyle='--')
plt.axvline(ub, linestyle='--');

plt.axvline(toxin_pA.std() - control_pA.std());
```

**6f.** (5 pts) Report whether or not you reject your null hypothesis **and why**. Do you think the toxin had a meaningful effect on the standard deviation of current fluctuations?

I do NOT reject the null hypothesis because the observed difference is within the 95% confidence interval of expected differences given the null hypothesis. The observed difference in the standard deviation of current fluctuations between control and toxin conditions is likely just due to random chance, and not do to any effect of the toxin.
